```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer


preprocess = ColumnTransformer(
    [
    ('num_scaler', StandardScaler(), numerical_data),
    ('nominal_encoder', OneHotEncoder(handle_unknown='ignore', drop='first'), nominal_data),
    ('ordinal_encoder', OrdinalEncoder(categories=[ ['Low', 'Medium', 'High'],['Tier 1', 'Tier 2', 'Tier 3']]), ordinal_data)
])

```

```python
preprocess
```

```python
model = make_pipeline(preprocess, LinearRegression())
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error


mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

print(f"mae : {mae}")
print(f"mse : {mse}")
print(f"rmse : {rmse}")
```
