The k-mean algorithms works as follow : 
1. specify the number of clusters K that you want to the data to be grouped into.
2. Randomly initialize K cluster centers or centroids. This can be done by randomly picking K data points to be the initial centroids
3. Assign each data points to the closest cluster centroid based on Euclidean distance. The data points closest to the given centriod are considered  part of that cluster
4. Recompute the cluster centriod by taking the mean of data points assigned to that clusters
5. Repeat the step 3 and 4 until the centroud stoip moviing or the iterations reac a specified limit. This is done ehaen the algorithms is converged
